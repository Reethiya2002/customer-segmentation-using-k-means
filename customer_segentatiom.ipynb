# import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# load data
df = pd.read_csv('mall_customers.csv')
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
# visualize anuual income vs spending score 
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
plt.xlabel('Anuual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('customer Distribution')
plt.show()
```

```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x
```

```python
x_scaler
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(x_scaler)
    wcss.append(kmeans.inertia_)    

print("wcss values:", wcss)

plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')    
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(x_scaler)
(df.head(10))
```

```python
df
```

```python
plt.figure(figsize=(8,6))
for cluster in range(5):
    plt.scatter(
        df[df['Cluster'] == cluster]['Annual Income (k$)'],
        df[df['Cluster'] == cluster]['Spending Score (1-100)'],
        label=f'Cluster {cluster}'  
    )
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')    
plt.title('Customer Segments')
plt.legend()
plt.show()    
```

```python
import joblib
# Save the model and scaler
joblib.dump(kmeans, 'kmeans_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
